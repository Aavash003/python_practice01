```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup("https://www.worldometers.info/coronavirus/","html.parser")
soup = requests.get(url)
soup
```

```python
#soup.text
```

```python
soup = BeautifulSoup(soup.text,"lxml") #converting to lxml
table_code = soup.table
table_code
```

```python
tags = table_code.find_all("tr")
tags
#print(type(tags))
#tags[-1].text
```

```python
data = []
for i in tags:
    y = i.text.split("\n")[1:]
    if y[0] != "":
        data.append(y)

data
```

```python
type(data[0])
```

```python
data[0]
```

```python
import csv
import pandas as pd

file = open("covid.csv","w")
x = csv.writer(file)
x.writerows(data)
file.close()
```

# Grouped Bar Chart

```python
import plotly.graph_objects as go
#animals=['giraffes', 'orangutans', 'monkeys']

fig = go.Figure(data=[
    go.Bar(name='TotalCases', x=df["Country,Other"][:10], y=df["TotalCases"][:10]),
    go.Bar(name='TotalDeaths', x=df["Country,Other"][:10], y=df["TotalDeaths"][:10])
])
# Change the bar mode
fig.update_layout(barmode='group')
fig.show()
```

# TotalCases vs TotalRecovered

```python
y2 = df["TotalRecovered"][:10]
y2_list = []
for values in y2:
    if type(values) == type(str()):
        t = values.replace(',',"")
        y2_list.append(float(t))
    else:
        y2_list.append(values)
```

```python
df["TotalRecovered"][:10] = y2_list
mean = df["TotalRecovered"][:10].mean()

df["TotalRecovered"][:10].fillna(mean,inplace=True)
df.head(10)
```

```python
mean
```

# Stacked Bar Chart

```python
import plotly.graph_objects as go
#animals=['giraffes', 'orangutans', 'monkeys']

fig = go.Figure(data=[
    go.Bar(name='TotalCases', x=df["Country,Other"][:10], y=df["TotalCases"][:10]),
    go.Bar(name='TotalDeaths', x=df["Country,Other"][:10], y=df["TotalRecovered"][:10])
])
# Change the bar mode
fig.update_layout(barmode='stack')
fig.show()
```
